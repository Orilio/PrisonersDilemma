```python
import networkx as nx
```

```python
G2 = nx.DiGraph()
G2.add_node('always cooperate')
G2.add_node('always defect')
G2.add_node('TitForTat')

G2.add_edge('always cooperate', 'always defect', score=29)
G2.add_edge('always defect', 'always cooperate', score=30)

G2.add_edge('always defect', 'TitForTat', score=20)
G2.add_edge('TitForTat', 'always defect', score=100)

G2.add_edge('TitForTat', 'always cooperate', score=100)
G2.add_edge('always cooperate', 'TitForTat', score=100)
```

```python
import matplotlib.pyplot as plt
# Re-attempt to visualize the directed graph with arrow directions based on score comparison, keeping both scores visible.

# Create a directed graph with adjusted arrow directions based on score comparison
G2_directed_adjusted = nx.DiGraph()

# Add nodes
G2_directed_adjusted.add_nodes_from(G2.nodes())

# Adjust edges to point from higher score to lower score
for u, v, data in G2.edges(data=True):
    score_uv = data['score']
    score_vu = G2[v][u]['score'] if G2.has_edge(v, u) else None

    # Determine edge direction based on score comparison
    if score_vu is not None and score_uv >= score_vu:
        G2_directed_adjusted.add_edge(u, v, score=(score_uv, score_vu))
    elif score_vu is not None:
        G2_directed_adjusted.add_edge(v, u, score=(score_vu, score_uv))
    else:
        G2_directed_adjusted.add_edge(u, v, score=(score_uv, None))

# Draw the adjusted directed graph with bidirectional scores displayed
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G2_directed_adjusted, seed=42)
nx.draw(G2_directed_adjusted, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True)
edge_labels_adjusted = nx.get_edge_attributes(G2_directed_adjusted, 'score')
edge_labels_adjusted_formatted = {key: f"{val[0]}/{val[1]}" if val[1] is not None else f"{val[0]}" for key, val in edge_labels_adjusted.items()}
nx.draw_networkx_edge_labels(G2_directed_adjusted, pos, edge_labels=edge_labels_adjusted_formatted, font_color="purple")

plt.title("Adjusted Directed Graph with Arrow Directions Based on Score Comparison")
plt.show()
```
